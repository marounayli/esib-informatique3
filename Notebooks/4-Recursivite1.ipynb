# Récursivité

Une approche à résoudre un problème en fonction des solutions d’instances plus petites du même problème.
Une fonction récursive est une fonction qui s'appelle elle-même au cours de son exécution.  
L’algorithme récursif est plus simple en écriture que les algorithmes itératifs surtout si le cas traité présente une structure récursive. 




L'exemple par execellence d'une fonction récursive est celui de la fonction factorielle d'un nombre entier qui présente naturellement une structure récursive :



Toute analyse récursive est composée essentiellement de 2 étapes :

- Détermination de la condition d’arrêt ainsi que des cas particuliers.
- Détermination du traitement dans le cas général.


```python
def factoriel(n):
    if n==0 or n==1:
        return 1
    return n*factoriel(n-1)

n=int(input("Saisir un entier : "))
print(n,"!=",factoriel(n),sep="")        
```

Mauvaise utilisation de la récursivité dans l'exemple qui suit : pas de condition d'arrêt

```python
def factoriel(n):
    return n*factoriel(n-1)

n=int(input("Saisir un entier : "))
print(n,"!=",factoriel(n),sep="")   
```

Exemple 2 : Somme des n premiers entiers 

```python
def somme(n):
    if n==1:
        return 1
    return n+somme(n-1)

n=int(input("Saisir un entier : "))
print("La somme des {} premiers entiers est {}".format(n,somme(n)))
```

Exemple 3 : Affichage de "n" Bonjours (Une fonction qui n'a pas de retour).

```python
def saluer(n):
    if n>=1:
        print("Bonjour")
        saluer(n-1)

n=int(input("Saisir un entier : "))
saluer(n)
```

Exemple 4 : conversion d'un entier positif du système décimal au système binaire (fonction sans retour)


```python
def binaire(n):
    if n>1:
       binaire(n//2)
    print(n%2,end="")
    

n=int(input("Saisir un entier : "))
binaire(n)    
```

Exemple 5: conversion d'un entier positif du système décimal au système binaire (fonction avec retour)


```python
def binaire2(n):
    if n==1:
        return "1"
    return binaire2(n//2)+str(n%2)

n=int(input("Saisir un entier : "))
print(binaire2(n))   
```
